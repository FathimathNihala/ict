```python
#3. Calculate the mean of each column of the dataset.

#means = iris_data.mean()
means = iris_data.mean(numeric_only=True)
print("\nMean of each columns:")
print(means)
```

```python
#4. Check for the null values present in the dataset.

null_values = iris_data.isnull().sum()
print("\nNull values in the dataset:")
print(null_values)
```

```python
#5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# Visualization 1: Boxplot for Sepal width by Classification
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SW', data=iris_data)
plt.title('Boxplot of Sepal Width by Classification')
plt.show()
```

```python
# Visualization 2: Scatter plot for prtals Length and Width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PL', y='PW', hue='Classification', data=iris_data)
plt.title('Scatter Plot for PETALS Length and Width')
plt.show()
```

```python
# Visualization 3: Bar chart for petal Width (PW)
plt.figure(figsize=(10, 6))
sns.barplot(x=iris_data['Classification'], y='PW', data=iris_data)
plt.title('Bar Chart for PETAL Width by Species')
plt.show()
```

```python
# Visualization : Bar chart for Sepal Width (SW)
plt.figure(figsize=(10, 6))
sns.barplot(x=iris_data['Classification'], y='SW', data=iris_data)
plt.title('Bar Chart for PETAL Width by Species')
plt.show()
```
